```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
distance, pulses = np.loadtxt("data.txt",unpack=True,delimiter=",",skiprows=1)
```

```python
plt.figure(1)
plt.plot(pulses,distance,"o")
plt.xlabel("Distance - [m]")
plt.ylabel("Pulses")
plt.show()
```

```python
fit = np.poly1d(np.polyfit(pulses,distance,1))
print(fit)
```
