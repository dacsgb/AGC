```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bit, angle_left = np.loadtxt("data.txt",unpack=True,delimiter=",",skiprows=1)
```

```python
plt.figure(1)
plt.plot(angle_left,bit,"o")
plt.ylabel("Bit Value")
plt.xlabel("Measured Angle")
plt.show()
```

```python
fit_left = np.poly1d(np.polyfit(angle_left,bit,1))
print(fit_left)
```
